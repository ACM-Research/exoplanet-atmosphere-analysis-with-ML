```python
#@title Set up datasets and pictures
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image
planet_result = pd.read_csv('results.csv')
planet_data = pd.read_csv('demo.csv')
```

```python
#@title See the lightwave of a planet! { vertical-output: true }
planet_input = "planet_three" #@param ["planet_one", "planet_two", "planet_three"]
```

```python
#@title See lightwave curve from planet
if(planet_input == "planet_one"):
    plt.plot(planet_data.iloc[0,1:])
    plt.show()

elif(planet_input == "planet_two"):
    plt.plot(planet_data.iloc[1,1:])
    plt.show()

else:
    plt.plot(planet_data.iloc[2,1:])
    plt.show()
```

```python
#@title Select the planet to analyze { vertical-output: true }
planet_show = "planet_three" #@param ["planet_one", "planet_two", "planet_three"]
```

```python
#@title See elemental composition from planet

planet_result.columns = ['Planet Number', 'CO2', 'H20', 'O2', 'N2', 'CH4']

if(planet_show == "planet_one"):
    print(planet_result.iloc[8135:8136,1:])
elif(planet_show == "planet_two"):
    print(planet_result.iloc[957:958,1:])
else:
    print(planet_result.iloc[3624:3625,1:])
```
